```python
data=pd.read_csv("electricity_price.csv")
data
```

```python
import matplotlib.pyplot as plt
from scipy import interpolate
x = np.arange(0, 10)
y = np.exp(-x/3.0)
f = interpolate.interp1d(x, y)

xnew = np.arange(0, 9, 0.1)
ynew = f(xnew)   # use interpolation function returned by `interp1d`
plt.plot(x, y, 'o', xnew, ynew, '-')
plt.show()
```

```python
x=list(data.index)
x=np.array(x)
y=np.array(data[data.columns[0]])

f = interpolate.interp1d(x, y)

xnew=np.arange(0, 24, 0.25)
ynew = f(xnew)
plt.plot(x, y, 'ro', xnew, ynew, 'b-')
```

```python
tck = interpolate.splrep(x,y)
xnew=np.arange(0, 24, 0.25)
ynew = interpolate.splev(xnew,tck,der=0)
plt.plot(x, y, 'ro', xnew, ynew, 'b-')
```

```python
x=list(data.index)
xnew=np.arange(0, 24, 0.25)
data2=pd.DataFrame(columns=data.columns)

for i in range(len(data.columns)):
    y=np.array(data[data.columns[i]])
    tck = interpolate.splrep(x,y)
    ynew = interpolate.splev(xnew,tck,der=0)
    data2[data2.columns[i]]=list(ynew)
```
