```python
import os
import pandas as pd
import numpy as np
from pylab import mpl,plt
```

```python
#获得目录下的所有文件
def get_file_name(file_dir):
    L=[]
    for root,dirs,files in os.walk(file_dir):
        for file in files:
    #print(file)
            L.append(file)
    return L
```

```python
#获得当前的文件目录
file_dir=os.getcwd()+"\\workload_data\\raw4"
print(file_dir)
files=[]
files=get_file_name(file_dir)
#print("get all file of the dir:")
print(files)
```

```python

workload_dataset=pd.DataFrame()

for  My_file_name in files:
    if My_file_name.find(".csv.gz")==-1:
        print(My_file_name)
        My_file_name=file_dir+"\\"+My_file_name
        data = pd.read_csv(My_file_name,header=None,engine='python')
        workload_dataset=pd.concat([workload_dataset,data])
    else:
        pass
```

```python
cols_name=["timestamp","priority","CPU_request","RAM_request"]
workload_dataset=workload_dataset[[0,8,9,10]]
workload_dataset.columns=cols_name
```
