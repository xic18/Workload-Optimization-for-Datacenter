```python
import os
import pandas as pd
import numpy as np
from pylab import mpl,plt
import random
```

```python
data=pd.read_excel("my_LMP.xlsx")
data
```

```python
import matplotlib.pyplot as plt
from scipy import interpolate
x=list(data.index)
xnew=np.arange(0, 24, 0.25)
data2=pd.DataFrame(columns=data.columns)

for i in range(len(data.columns)):
    y=np.array(data[data.columns[i]])
    tck = interpolate.splrep(x,y)
    ynew = interpolate.splev(xnew,tck,der=0)
    data2[data2.columns[i]]=list(ynew)
```

```python
data2.plot()
plt.show()
```

```python
data2.to_csv("my_LMP_spline.csv")
```
