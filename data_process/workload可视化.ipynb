```python
import os
import pandas as pd
import numpy as np
from pylab import mpl,plt
```

```python
df1=pd.read_csv("load_info.csv")
df2=pd.read_csv("load_info2.csv")
df3=pd.read_csv("load_info3.csv")
df4=pd.read_csv("load_info4.csv")
```

```python
df=pd.concat([df1,df2,df3,df4])
df.index=list(range(len(df)))
df
```

```python
#["time","num_0","CPU_usage_0","RAM_usage_0","num_1","CPU_usage_1","RAM_usage_1","num_2","CPU_usage_2","RAM_usage_2"]
df[["num_0","num_1","num_2"]].plot()
df[["CPU_usage_0","CPU_usage_1","CPU_usage_2",]].plot()
df[["RAM_usage_0","RAM_usage_1","RAM_usage_2",]].plot()
plt.show()
```

```python
df["CPU"]=df["CPU_usage_0"]+df["CPU_usage_1"]+df["CPU_usage_2"]
df["RAM"]=df["RAM_usage_0"]+df["RAM_usage_1"]+df["RAM_usage_2"]
df[["CPU","RAM"]].plot()
plt.show()
```

```python
df["CPU_usage_0"]=df["CPU_usage_0"]/12000
df["CPU_usage_1"]=df["CPU_usage_1"]/12000
df["CPU_usage_2"]=df["CPU_usage_2"]/12000
df["RAM_usage_0"]=df["RAM_usage_0"]/12000
df["RAM_usage_1"]=df["RAM_usage_1"]/12000
df["RAM_usage_2"]=df["RAM_usage_2"]/12000
df[["num_0","num_1","num_2"]].plot()
df[["CPU_usage_0","CPU_usage_1","CPU_usage_2",]].plot()
df[["RAM_usage_0","RAM_usage_1","RAM_usage_2",]].plot()
plt.show()
```
